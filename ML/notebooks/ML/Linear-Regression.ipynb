```python
import pandas as pd

# Create a dictionary containing our data
data = {'Study_Hours': [1, 2, 3, 4, 5],
        'Test_Marks': [55, 60, 78, 85, 92]}

# Create a Pandas DataFrame from the dictionary
df = pd.DataFrame(data)

# Print the DataFrame to see our data
print(df)
```

```python
import matplotlib.pyplot as plt

# Create a scatter plot
plt.scatter(df['Study_Hours'], df['Test_Marks'])

# Add labels and a title
plt.xlabel('Study Hours')
plt.ylabel('Test Marks')
plt.title('Relationship between Study Hours and Test Marks')

# Show the plot
plt.show()
```
