```python
df = pd.read_csv("Titanic-Dataset.csv")
df.head()
```

```python
sns.countplot(data= df , x= "Survived")
plt.title("Passenger Survival count ")
plt.show()
```

```python
sns.countplot(data=df , x="Sex")
plt.title("Sex ratio")
plt.show()
```

```python
sns.histplot(data=df , x="Age" , bins=200 , kde=True)
plt.xlabel("Ages")
plt.ylabel("Number of People")
plt.show()
```

```python
df.head()
```

```python
sns.countplot(data= df , x= "Pclass" , hue="Survived")
```

```python
sns.set_theme(style="darkgrid") # Example theme
```

```python
sns.countplot(data=df, x='Sex', hue='Survived', palette='viridis') # Added palette

# Add labels and title (important for customized plots)
plt.title('Survival Count by Sex (Viridis Palette)')
plt.xlabel('Sex')
plt.ylabel('Number of Passengers')
plt.legend(title='Survived', labels=['No', 'Yes']) # Keep legend clear
```

```python
plt.figure(figsize=(10 , 6))
sns.histplot(data=df , x="Age" ,color="green")
plt.title("Distribution of Age")
plt.show()
```

```python


# Set figure size
plt.figure(figsize=(10, 7))

# CREATE THE BOXPLOT: Compare 'Age' across 'Pclass'
sns.boxplot(data=df, x='Pclass', y='Age', palette='Set2') # Example palette

# Add labels and title
plt.title('Age Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')

#show
plt.show()
```

```python
sns.boxplot(data=df , x = "Age")
```

```python
sns.violinplot(data=df, x='Pclass', y='Age', palette='viridis', inner='quartile') # Added inner='quartile'

# Add labels and title
plt.title('Age Distribution by Passenger Class (Violin Plot)')
plt.xlabel('Passenger Class')
plt.ylabel('Age')


plt.show()
```
